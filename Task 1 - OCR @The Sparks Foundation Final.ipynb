## Import all required python libraries

```python
import cv2
import matplotlib.pyplot as plt
import easyocr
from pylab import rcParams
import warnings
warnings.filterwarnings("ignore")
from IPython.display import Image
rcParams['figure.figsize'] = 10,20
```

## Download the language model - English

```python
Lang_reader = easyocr.Reader(['en'])
```

## Image Preview For OCR

```python
Image(r'..\Img\Test_img.jpg')
```

## Text Detection In Test_img.jpg

```python
output1 = Lang_reader.readtext(r'..\Img\Test_img.jpg')
```

## Print the detected texts in image

### List Of Bounding box co-ordinates, Text and Their Confidence level

```python
output1
```

## Extract the text & its bounding box coordinates

```python
cord1 = output1[-3][0]
print(output1[-3][1])
```

## Minimum value of coordinates

```python
x_min1,y_min1 = [min(idx) for idx in zip(*cord1)]
```

## Maximum value of coordinates

```python
x_max1,y_max1 = [max(idx) for idx in zip(*cord1)]
```

## Visuals of Bounding box co-ordinates of particular text In Image

```python
image = cv2.imread('..\Img\Test_img.jpg')
cv2.rectangle(image,(int(x_min1),int(y_min1)),(int(x_max1),int(y_max1)),(0,255,0),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
